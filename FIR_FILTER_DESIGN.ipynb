```python
#Transfer function of FIR Filter
def fir(order,w):
  sum = 0
  # sum = (b[0]*np.exp(complex(0,0*w))+b[1]*np.exp(complex(0,(-1)*w))+b[2]*np.exp(complex(0,(-2)*w))+b[3]*np.exp(complex(0,(-3)*w))+b[4] *np.exp(complex(0,(-4)*w))+b[5]*np.exp(complex(0,(-5)*w)))
  for i in range(len(b)):
    sum = sum + b[i] * np.exp(complex(0,(-1)*(i)*w))
  return sum
```

```python
#Array of frequency
w1 = np.arange(0,24000)
print(len(w1))
```

```python
len(w1)
w2 = [i*2*np.pi/48000 for i in w1]
print(w2)
```

```python
result = []
for i in range(len(w2)):
  result.append(fir(5,w2[i]))
```
