```python
#Bartlet window
w3 = []
N = 4
for n in range(int(N/2) + 1):
  w3.append(2*n/N)
for n in range(int(N/2)+1,N+1):
  w3.append((2-((2*n)/N)))
w3
```

```python
h3 = []
tau = (N-1)/2
print(tau)
w_c = 0.65 * np.pi
print(w_c)
for n in range(N):
  h3.append((np.sin(0.65*np.pi * (n-tau)))/(np.pi*(n-tau)))
h3.insert(2,0.65)
h3
```

```python
hd3 = []
for i in range(N+1):
  hd3.append(h3[i]*w3[i])
```

```python
hd3
```

```python
h3_dft = np.fft.fft(hd3,200)
```

```python
y3_dft = x_dft*h3_dft
y3_final = np.fft.ifft(y3_dft)
```

```python
x_range = np.arange(200)
# plt.plot(x_range,y_final)
plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
plt.plot(x_axis,w3)
plt.title("Bartlet Window")
plt.subplot(3,2,3)
plt.plot(x_axis,h3)
plt.title("Coefficients")
plt.subplot(3,2,5)
plt.plot(x_axis,hd3)
plt.title("Filter")

plt.subplot(3,2,2)
plt.plot(x_range,x)
plt.title("Signal")
plt.subplot(3,2,4)
plt.plot(x_axis,hd3)
plt.title("Filter")
plt.subplot(3,2,6)
plt.plot(x_range,y3_final)
plt.title("Filtered Signal")
```
